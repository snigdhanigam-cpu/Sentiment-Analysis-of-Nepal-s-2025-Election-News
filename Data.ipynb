```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
import string
# Remove punctuation
df1['Raw Text'] = df1['Raw Text'].str.replace(f"[{string.punctuation}]", '', regex=True)
# Lowercase
df1['Raw Text'] = df1['Raw Text'].str.lower()
# Remove stop words - note x.split() to iterate over WORDS not characters
stop_words = set(stopwords.words('english'))
df1['Raw Text'] = df1['Raw Text'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
print(df1['Raw Text'])
```

```python
#tokenization
df1['Tokenized_Text'] = df1['Raw Text'].apply(word_tokenize)
```
