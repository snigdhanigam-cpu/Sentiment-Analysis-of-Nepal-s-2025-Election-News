```python
import pandas as pd
df3 = pd.read_csv("newspaper_articles.csv")
df4 = pd.read_csv("bs4_advanced_articles.csv")
```

```python
df3 = df3.dropna(subset=["Raw Text"])
df4 = df4.dropna(subset=["text"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Combine both datasets to fit the vectorizer on all text
all_texts = list(df3['Raw Text']) + list(df4['text'])

# Fit TF-IDF on all texts
vectorizer = TfidfVectorizer()
vectorizer.fit(all_texts)

# Transform each dataset separately
matrix1 = vectorizer.transform(df3['Raw Text'])
matrix2 = vectorizer.transform(df4['text'])

# Calculate cosine similarity between the two matrices
similarity_matrix = cosine_similarity(matrix1, matrix2)

# For each row in df1, find the max similarity score against all of df2
df3['max_similarity'] = similarity_matrix.max(axis=1)

# Filter to only keep highly similar texts (adjust threshold as needed)
threshold = 0.60
highly_similar = df3[df3['max_similarity'] >= threshold]
highly_similar.to_csv('highly_similar_articles.csv', index=False)
```
